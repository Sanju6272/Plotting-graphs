# Plot Job_Year
Problem Statement :
Plot the line graph between no. of Job postings with respect to year.
Print the year and the number of job posting as integer value.
Note: Year should be in ascending order.
Output Format :
year1 job_posting1
year2 job_posting2
. . . 
. . .
. . .

```python
import matplotlib.pyplot as plt
import numpy as np
import csv

with open('amazon_jobs_dataset.csv',encoding='utf8') as file_obj:
    file_data=csv.DictReader(file_obj)
    
    d={}
    
    for row in file_data:
        if((row['Posting_date'].split())[2] in d):
            d[row['Posting_date'].split()[2]]=d[(row['Posting_date'].split())[2]]+1
        else:
            d[(row['Posting_date'].split())[2]]=1
        
year=[]
jobs=[]

for i in d:
    year.append(int(i))
    jobs.append(d[i])

plt.scatter(year,jobs)
plt.plot(year,jobs)
plt.show()


for i in sorted(d.keys()):
    print(i, d[i])
```

# Plot Month_Job
Problem Statement :
Plot the Bar graph between Month vs Job Openings.
Print the month name and the number of job posting as integer value.
Order of months doesn't matter.
Output Format :
month_name1 job_posting1
month_name2 job_posting2
. . . 
. . .
. . .

```python
import csv
import matplotlib.pyplot as plt
import numpy as np

with open('amazon_jobs_dataset.csv',encoding='utf8') as file_obj:
    file_data=csv.DictReader(file_obj,skipinitialspace=True)
    
    d={}
    
    for row in file_data:
        a=row['Posting_date'].split()[0]
        if a in d:
            d[a]=d[a]+1
        else:
            d[a]=1
            
    month=[]
    jobs=[]
    
    for i in d:
        month.append(i)
        jobs.append(d[i])
        
    plt.bar(month,jobs)
    plt.xticks(rotation=90)
    plt.show()
            
for i in d:
    print(i,d[i])
    
```

# Plot city_job
Problem Statement :
Plot the Pie chart between Indian cities vs No. of jobs opening.
Print the Indian cities and %age of Job distribution in India upto 2 decimal places.
Note: %age of Job distribution should be in descending order.
Output Format :
city1 percentage1
city2 percentage2
. . . 
. . .
. . .


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import csv

with open('amazon_jobs_dataset.csv',encoding='ISO-8859-1') as file_obj:
    file_data=csv.DictReader(file_obj,skipinitialspace=True)
    dict={}
    indiacitylist=[]
    
    for row in file_data:
        location =row['location']
        s=location.split(" ")
        for i in s:
            if i=="" or i==" ":
                s.remove(i)
        if 'IN,' in s:
            indiacitylist.append(s[2])
            
for city in indiacitylist:
    dict[city]=dict.get(city,0)+1
    
citys=[]
values=[]

for key in dict:
    citys.append(key)
    values.append(dict[key])
 
labels=citys
plt.pie(values,labels=labels,autopct="%.2f%%")

sum=0
for key in dict:
    sum+=dict[key]

for key in dict:
    dict[key]=(dict[key]/sum)*100
    dict[key]=format(dict[key], ".2f")
    
for key in dict:
    print(key,dict[key])
plt.show()
```

# Plot Job_Java
Problem Statement :
Plot the scatter graph between year vs No. of jobs opening related to Java.
Print the year and number of Jobs opening in Java Profile.
Note: Use the Keyword 'Java' or 'java' in Basic Qualification feature for finding the job opening related to Java Profile. Print the year in ascending order.
Output Format :
year1 JobOpening1
year2 JobOpening2
. . . 
. . .
. . .


```python
import csv
import matplotlib.pyplot as plt
import numpy as np

with open('amazon_jobs_dataset.csv',encoding='utf8') as file_obj:
    file_data=csv.DictReader(file_obj,skipinitialspace=True)
    
    y=[]
    
    for row in file_data:
        if 'java' in row['BASIC QUALIFICATIONS'] or 'Java' in row['BASIC QUALIFICATIONS']:
            y.append(row['Posting_date'][len(row['Posting_date'])-4:])
  
d={}
for i in y:
    if i in d:
        d[i]+=1
    else:
        d[i]=1

year=[]
jobs=[]


for i in d:
    year.append(i)
    jobs.append(d[i])
    
np_year=np.array(year,dtype='int64')
np_jobs=np.array(jobs,dtype='int64')

np_jobs=np_jobs[np.argsort(np_year)]
np_year=np.sort(np_year)

plt.scatter(np_year,np_jobs)
plt.show()

for i in range(len(np_year)):
    print(np_year[i], np_jobs[i])
```
