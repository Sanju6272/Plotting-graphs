```python
import matplotlib.pyplot as plt
import numpy as np
x=np.arange(0,5,0.1)
y=x**3
y2=x**2
plt.plot(x,y,color='red',label='x^3')
plt.plot(x,y2,color='blue',label='y^2')
plt.legend()
plt.xlabel('x')
plt.ylabel('y')
plt.text(3,40,'sanju',fontsize=12)
plt.grid()
plt.show()
```

```python
labels=['a','b','c','d']
sizes=[3,4,5,2]
colors=['orange','blue','red','pink']
explode=[0.1,0,0,0]
plt.pie(sizes,colors=colors,explode=explode,labels=labels,autopct='%.2f%%',startangle=120)
plt.axis('equal')
plt.title('split')
plt.show()
```

```python
import matplotlib.pyplot as plt

company=['HP','Dell','Lenovo','Asus','Apple','Acer']
sold=[20000,43000,15000,17000,22000,13000]

plt.pie(sold,labels=company,autopct='%.1f')
plt.axis('equal')
plt.show()
for i in range(6):
    a=(sold[i]/sum(sold))*100
    print(company[i],'{:.1f}'.format(a) )
```
